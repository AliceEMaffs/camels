```python
def plot_parameter_variations(simulations, category, redshift_info):
    """Create grid plot showing parameter variations for all simulations."""
    config = get_config(dataset="1P", simulation=simulations[0])
    
    # Set up parameters and variations
    params = range(1, 7)  # p1 through p6
    param_labels = ['Ωm', 'σ8', 'ASN1', 'AAGN1', 'ASN2', 'AAGN2']
    variations = ['n2', 'n1', '0', '1', '2']
    
    # Use viridis colormap for better color scheme
    colors = plt.cm.viridis(np.linspace(0, 1, len(variations)))
    
    # Create figure with minimal spacing
    fig, axes = plt.subplots(len(simulations), 6, figsize=(24, 20))
    plt.subplots_adjust(hspace=0.1, wspace=0.1, left=0.06, right=0.98, top=0.95, bottom=0.06)
    
    for sim_idx, simulation in enumerate(simulations):
        sim_config = get_config(dataset="1P", simulation=simulation)
        
        for param_idx, (param, param_label) in enumerate(zip(params, param_labels)):
            ax = axes[sim_idx, param_idx]
            
            for var_idx, var in enumerate(variations):
                try:
                    filter_system = get_safe_name("GALEX FUV", filter_system_only=True)
                    data_dir = os.path.join(sim_config["lf_data_dir"][category][filter_system],
                                          get_safe_name(redshift_info['label']))
                    filename = f"UVLF_1P_p{param}_{var}_GALEX_FUV_{redshift_info['label']}_{category}.txt"
                    filepath = os.path.join(data_dir, filename)
                    
                    if os.path.exists(filepath):
                        data = pd.read_csv(filepath, sep='\t')
                        ax.errorbar(data['magnitude'], 
                                  data['phi'],
                                  yerr=data['phi_sigma'],
                                  fmt='o-', color=colors[var_idx],
                                  markersize=4, capsize=3, alpha=0.7,
                                  label=f'{var}σ' if var != '0' else 'fiducial')
                
                except Exception as e:
                    print(f"Error processing {filepath}: {str(e)}")
            
            ax.grid(True, alpha=0.3, linestyle='--')
            ax.set_xlim(-23, -14.5)
            ax.set_ylim(-5, -2)
            
            # Only show x-label for bottom row
            if sim_idx == len(simulations)-1:
                ax.set_xlabel('M$_{\\rm{AB}}$', fontsize=10)
            else:
                ax.set_xticklabels([])
            
            # Only show y-label for first column
            if param_idx == 0:
                ax.set_ylabel('log$_{10}$ φ [Mpc$^{-3}$ mag$^{-1}$]', fontsize=10)
                ax.text(-0.3, 0.5, simulation, transform=ax.transAxes,
                       rotation=90, va='center', ha='center', fontsize=12)
            else:
                ax.set_yticklabels([])
            
            # Only show title for top row
            if sim_idx == 0:
                ax.set_title(param_label, fontsize=12, pad=2)
            
            # Legend only in top-left plot
            if sim_idx == 0 and param_idx == 0:
                leg = ax.legend(fontsize=8, ncol=1, bbox_to_anchor=(1.0, 1.1))
                for line in leg.get_lines():
                    line.set_alpha(1.0)  # Make legend lines fully opaque
    
    plt.suptitle(f'UV Luminosity Functions ({category}, z={redshift_info["redshift"]})',
                fontsize=14, y=0.98)
    
    return fig

# Similarly update plot_single_simulation_variations with the same color scheme
def plot_single_simulation_variations(simulation, category, redshift_info):
    """Plot parameter variations for a single simulation."""
    config = get_config(dataset="1P", simulation=simulation)
    
    params = range(1, 7)
    param_labels = ['Ωm', 'σ8', 'ASN1', 'AAGN1', 'ASN2', 'AAGN2']
    variations = ['n2', 'n1', '0', '1', '2']
    colors = plt.cm.viridis(np.linspace(0, 1, len(variations)))
    
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    plt.subplots_adjust(hspace=0.2, wspace=0.15, left=0.1, right=0.95, top=0.9, bottom=0.1)
    axes = axes.flatten()
    
    for param_idx, (param, param_label) in enumerate(zip(params, param_labels)):
        ax = axes[param_idx]
        
        for var_idx, var in enumerate(variations):
            filter_system = get_safe_name("GALEX FUV", filter_system_only=True)
            data_dir = os.path.join(config["lf_data_dir"][category][filter_system],
                                  get_safe_name(redshift_info['label']))
            filename = f"UVLF_1P_p{param}_{var}_GALEX_FUV_{redshift_info['label']}_{category}.txt"
            filepath = os.path.join(data_dir, filename)
            
            try:
                if os.path.exists(filepath):
                    data = pd.read_csv(filepath, sep='\t')
                    ax.errorbar(data['magnitude'], 
                              data['phi'],
                              yerr=data['phi_sigma'],
                              fmt='o-', color=colors[var_idx],
                              markersize=4, capsize=3, alpha=0.7,
                              label=f'{var}σ' if var != '0' else 'fiducial')
            except Exception as e:
                print(f"Error processing {filepath}: {str(e)}")
        
        ax.grid(True, alpha=0.3, linestyle='--')
        ax.set_xlim(-23, -14.5)
        ax.set_ylim(-5, -2)
        ax.set_title(param_label, fontsize=12)
        
        if param_idx >= 3:
            ax.set_xlabel('M$_{\\rm{AB}}$', fontsize=10)
        if param_idx % 3 == 0:
            ax.set_ylabel('log$_{10}$ φ [Mpc$^{-3}$ mag$^{-1}$]', fontsize=10)
        if param_idx == 0:
            ax.legend(fontsize=8, ncol=1)
    
    plt.suptitle(f'{simulation} UV Luminosity Functions\n({category}, z={redshift_info["redshift"]})',
                fontsize=14)
    
    return fig
```

```python
def plot_parameter_variations_cbar(simulations, category, redshift_info):
    """Create grid plot showing parameter variations for all simulations."""
    config = get_config(dataset="1P", simulation=simulations[0])
    
    params = range(1, 7)
    param_labels = ['Ωm', 'σ8', 'ASN1', 'AAGN1', 'ASN2', 'AAGN2']
    variations = ['n2', 'n1', '0', '1', '2']
    
    # Create figure with parameters as rows and simulations as columns
    fig, axes = plt.subplots(6, len(simulations), figsize=(20, 15))
    plt.subplots_adjust(hspace=0.1, wspace=0.15, left=0.08, right=0.92, top=0.95, bottom=0.08)
    
    # Create custom colormap
    cmap = plt.cm.viridis
    norm = mpl.colors.BoundaryNorm([-2, -1, 0, 1, 2], cmap.N)
    colors = cmap(np.linspace(0, 1, len(variations)))
    
    for param_idx, (param, param_label) in enumerate(zip(params, param_labels)):
        for sim_idx, simulation in enumerate(simulations):
            sim_config = get_config(dataset="1P", simulation=simulation)
            ax = axes[param_idx, sim_idx]
            
            for var_idx, var in enumerate(variations):
                try:
                    filter_system = get_safe_name("GALEX FUV", filter_system_only=True)
                    data_dir = os.path.join(sim_config["lf_data_dir"][category][filter_system],
                                          get_safe_name(redshift_info['label']))
                    filename = f"UVLF_1P_p{param}_{var}_GALEX_FUV_{redshift_info['label']}_{category}.txt"
                    filepath = os.path.join(data_dir, filename)
                    
                    if os.path.exists(filepath):
                        data = pd.read_csv(filepath, sep='\t')
                        ax.errorbar(data['magnitude'], 
                                  data['phi'],
                                  yerr=data['phi_sigma'],
                                  fmt='o-', color=colors[var_idx],
                                  markersize=4, capsize=3, alpha=0.7)
                
                except Exception as e:
                    print(f"Error processing {filepath}: {str(e)}")
            
            ax.grid(True, alpha=0.3, linestyle='--')
            ax.set_xlim(-24, -14.5)
            ax.set_ylim(-5, -1.5)
            
            if param_idx == len(params)-1:
                ax.set_xlabel('M$_{\\rm{AB}}$', fontsize=10)
            else:
                ax.set_xticklabels([])
            
            if sim_idx == 0:
                ax.set_ylabel(f'{param_label}\nlog$_{10}$ φ [Mpc$^{-3}$ mag$^{-1}$]', fontsize=10)
            else:
                ax.set_yticklabels([])
            
            if param_idx == 0:
                ax.set_title(simulation, fontsize=12, pad=2)
    
    # Add colorbar
    cbar_ax = fig.add_axes([0.93, 0.15, 0.02, 0.7])  # [left, bottom, width, height]
    sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
    sm.set_array([])
    cbar = fig.colorbar(sm, cax=cbar_ax, ticks=[-2, -1, 0, 1, 2])
    cbar.set_label('Parameter Variation (σ)', rotation=270, labelpad=15)
    cbar.ax.set_yticklabels(['n2', 'n1', 'fiducial', '1', '2'])
    
    plt.suptitle(f'UV Luminosity Functions ({category}, z={redshift_info["redshift"]})',
                fontsize=14, y=0.98)
    
    return fig
```
