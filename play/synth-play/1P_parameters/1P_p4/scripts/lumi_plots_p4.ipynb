```python
# Python script to plot magi.qdp

import matplotlib.pyplot as plt
from matplotlib import rc

# Open the .qdp file
def get_arnouts(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()

    # Initialize lists to store x and y values
    arn_x = []
    arn_y = []

    # Read data between 'no no no' lines
    start_index = lines.index(' no no no no\n') + 1
    end_index = lines.index(' no no no no\n', start_index)

    for line in lines[start_index:end_index]:
        # Split the line and extract x and y values
        parts = line.strip().split()
        arn_x.append(float(parts[0]))
        arn_y.append(float(parts[1]))

    return arn_x, arn_y


arn_x, arn_y = get_arnouts('/home/jovyan/camels/play/synth-play/1P_parameters/1P_p4/outfiles/magi_1P_p4_1.qdp')
```

```python
# Python script to plot magi.qdp

import matplotlib.pyplot as plt
# XMM
def get_data_from_magi(filename):
    

    # Open the .qdp file
    with open(filename, 'r') as file:
        lines = file.readlines()

    # Initialize lists to store x and y values
    x = []
    y = []

    # Find the index of the start and end lines
    start_index = lines.index('   -22.00000  no no no\n') + 1
    end_index = lines.index(' no no no no\n', start_index)

    # Read data between start and end lines
    for line in lines[start_index:end_index]:
        # Split the line and extract x and y values
        parts = line.strip().split()
        x.append(float(parts[0]))
        y.append(float(parts[1]))
        
    return x, y

x_p4_1,y_p4_1 = get_data_from_magi('/home/jovyan/camels/play/synth-play/1P_parameters/1P_p4/outfiles/magi_1P_p4_1.qdp')
x_p4_2,y_p4_2= get_data_from_magi('/home/jovyan/camels/play/synth-play/1P_parameters/1P_p4/outfiles/magi_1P_p4_2.qdp')
x_p4_n1,y_p4_n1 = get_data_from_magi('/home/jovyan/camels/play/synth-play/1P_parameters/1P_p4/outfiles/magi_1P_p4_n1.qdp')
x_p4_n2,y_p4_n2 = get_data_from_magi('/home/jovyan/camels/play/synth-play/1P_parameters/1P_p4/outfiles/magi_1P_p4_n2.qdp')
```
