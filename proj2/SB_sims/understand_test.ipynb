```python
# read in SB set and have a look at the data.
import h5py

def print_hdf5_structure(file_path):
    """Print the complete structure of an HDF5 file with indentation"""
    def print_item(name, obj, indent=0):
        indent_str = "    " * indent  # 4 spaces per indent level
        if isinstance(obj, h5py.Dataset):
            print(f"{indent_str}Dataset: {name}, Shape: {obj.shape}, Data Type: {obj.dtype}")
        elif isinstance(obj, h5py.Group):
            print(f"{indent_str}Group: {name}")

    with h5py.File(file_path, 'r') as f:
        f.visititems(lambda name, obj: print_item(name, obj, name.count('/')))

# Use the function
file_path = "/disk/xray15/aem2/data/28pams/IllustrisTNG/SB/photometry/alice_galex.h5"

print("\nHDF5 File Structure:")
print("-" * 50)
print_hdf5_structure(file_path)
print("-" * 50)
```
