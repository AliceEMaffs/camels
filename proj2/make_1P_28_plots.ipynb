```python

# Comprehensive parameter configuration
param_config = {
    # Cosmological parameters
    "Omega0": {
        "label": r"$\Omega_0$",
        "group": "cosmo",
        "color": "royalblue",
        "index": 1
    },
    "sigma8": {
        "label": r"$\sigma_8$",
        "group": "cosmo", 
        "color": "royalblue",
        "index": 2
    },
    "WindEnergyIn1e51erg": {
        "label": r"$E_\mathrm{wind,1e51}$",
        "group": "winds",
        "color": "forestgreen",
        "index": 3
    },
    "RadioFeedbackFactor": {
        "label": r"$f_\mathrm{radio,1}$",
        "group": "agn_feedback",
        "color": "crimson",
        "index": 4
    },
    "VariableWindVelFactor": {
        "label": r"$v_\mathrm{wind,var}$",
        "group": "winds",
        "color": "forestgreen",
        "index": 5
    },
    "RadioFeedbackReiorientationFactor": {
        "label": r"$f_\mathrm{radio,reor}$",
        "group": "agn_feedback",
        "color": "crimson",
        "index": 6
    },
    "OmegaBaryon": {
        "label": r"$\Omega_\mathrm{b}$",
        "group": "cosmo",
        "color": "royalblue",
        "index": 7
    },
    "HubbleParam": {
        "label": r"$h$",
        "group": "cosmo",
        "color": "royalblue",
        "index": 8
    },
    "n_s": {
        "label": r"$n_\mathrm{s}$",
        "group": "cosmo",
        "color": "royalblue",
        "index": 9
    },
    "MaxSfrTimescale": {
        "label": r"$t_\mathrm{SF,max}$",
        "group": "sf_ism",
        "color": "teal",
        "index": 10
    },
    "FactorForSofterEQS": {
        "label": r"$f_\mathrm{EQS}$",
        "group": "sf_ism",
        "color": "teal",
        "index": 11
    },
    "IMFslope": {
        "label": r"$\alpha_\mathrm{IMF}$",
        "group": "stellar_pop",
        "color": "indigo",
        "index": 12
    },
    "SNII_MinMass_Msun": {
        "label": r"$M_\mathrm{SNII,min}$",
        "group": "stellar_pop",
        "color": "indigo",
        "index": 13
    },
    "ThermalWindFraction": {
        "label": r"$f_\mathrm{therm}$",
        "group": "winds",
        "color": "forestgreen",
        "index": 14
    },
    "VariableWindSpecMomentum": {
        "label": r"$p_\mathrm{wind,spec}$",
        "group": "winds",
        "color": "forestgreen",
        "index": 15
    },
    "WindFreeTravelDensFac": {
        "label": r"$n_\mathrm{wind,free}$",
        "group": "winds",
        "color": "forestgreen",
        "index": 16
    },
    "MinWindVel": {
        "label": r"$v_\mathrm{wind,min}$",
        "group": "winds",
        "color": "forestgreen",
        "index": 17
    },
    "WindEnergyReductionFactor": {
        "label": r"$f_\mathrm{wind,E}$",
        "group": "winds",
        "color": "forestgreen",
        "index": 18
    },
    "WindEnergyReductionMetallicity": {
        "label": r"$Z_\mathrm{wind,E}$",
        "group": "winds",
        "color": "forestgreen",
        "index": 19
    },
    "WindEnergyReductionExponent": {
        "label": r"$\alpha_\mathrm{wind,E}$",
        "group": "winds",
        "color": "forestgreen",
        "index": 20
    },
    "WindDumpFactor": {
        "label": r"$f_\mathrm{dump}$",
        "group": "winds",
        "color": "forestgreen",
        "index": 21
    },
    "SeedBlackHoleMass": {
        "label": r"$M_\mathrm{BH,seed}$",
        "group": "bh_growth",
        "color": "darkorange",
        "index": 22
    },
    "BlackHoleAccretionFactor": {
        "label": r"$f_\mathrm{BH,Bondi}$",
        "group": "bh_growth",
        "color": "darkorange",
        "index": 23
    },
    "BlackHoleEddingtonFactor": {
        "label": r"$f_\mathrm{BH,Edd}$",
        "group": "bh_growth",
        "color": "darkorange",
        "index": 24
    },
    "BlackHoleFeedbackFactor": {
        "label": r"$f_\mathrm{BH,FB}$",
        "group": "agn_feedback",
        "color": "crimson",
        "index": 25
    },
    "BlackHoleRadiativeEfficiency": {
        "label": r"$\epsilon_\mathrm{rad}$",
        "group": "agn_feedback",
        "color": "crimson",
        "index": 26
    },
    "QuasarThreshold": {
        "label": r"$L_\mathrm{Edd}$",
        "group": "agn_feedback",
        "color": "crimson",
        "index": 27
    },
    "QuasarThresholdPower": {
        "label": r"$\alpha_\mathrm{QSO}$",
        "group": "agn_feedback",
        "color": "crimson",
        "index": 28
    },
    # # Special Parameters - these don't correspond directly to parameter indices
    # "A_SN1": {
    #     "label": r"$A_\mathrm{SN1}$",
    #     "group": "sn_special",
    #     "color": "forestgreen"
    # },
    # "A_SN2": {
    #     "label": r"$A_\mathrm{SN2}$",
    #     "group": "sn_special",
    #     "color": "forestgreen"
    # },
    # "A_AGN1": {
    #     "label": r"$A_\mathrm{AGN1}$",
    #     "group": "agn_special",
    #     "color": "crimson"
    # },
    # "A_AGN2": {
    #     "label": r"$A_\mathrm{AGN2}$",
    #     "group": "agn_special",
    #     "color": "crimson"
    # }
}

def get_config(dataset="1P", simulation="IllustrisTNG"):
    """Get configuration for a specific dataset and simulation."""
    # Get the complete configuration from variables_config
    config_vars = get_variable_config(dataset=dataset, simulation=simulation)
    
    # Create our configuration using the imported values
    config = {
        "base_dir": config_vars["base_dir"],
        "simulation": simulation,
        "redshift_values": config_vars["redshift_values"],
        "lf_data_dir": config_vars["lf_data_dir"],
        "plots_dir": config_vars["plots_dir"],
        "uvlf_limits": config_vars["uvlf_limits"],
        "n_bins_lf": config_vars["n_bins_lf"],
        "colour_limits": config_vars["colour_limits"],
        "n_bins_colour": config_vars["n_bins_colour"],
        "colour_data_dir": config_vars["colour_data_dir"],
        "filters": config_vars["filters"],
        "colour_pairs": config_vars["colour_pairs"]
    }
    
    return config

# Create a mapping from parameter index to name
param_idx_to_name = {}
for param_name, param_info in param_config.items():
    if "index" in param_info:
        param_idx_to_name[param_info["index"]] = param_name

def read_parameter_values(dataset="1P", simulation="IllustrisTNG"):
    """
    Read parameter values from the parameter file using the indices 
    defined in param_config.
    
    Args:
        dataset: Dataset name (default: "1P")
        simulation: Simulation name (default: "IllustrisTNG")
        
    Returns:
        Dictionary mapping parameter indices to variation values
    """
    # # Hardcoded parameter values as fallback
    # hardcoded_values = {
    #     # Omega0 (param 1)
    #     1: {
    #         'n2': {'Omega0': '0.1'},
    #         'n1': {'Omega0': '0.2'},
    #         '0': {'Omega0': '0.3'},
    #         '1': {'Omega0': '0.4'},
    #         '2': {'Omega0': '0.5'}
    #     },
    #     # sigma8 (param 2)
    #     2: {
    #         'n2': {'sigma8': '0.6'},
    #         'n1': {'sigma8': '0.7'},
    #         '0': {'sigma8': '0.8'},
    #         '1': {'sigma8': '0.9'},
    #         '2': {'sigma8': '1.0'}
    #     },
    #     # WindEnergyIn1e51erg (param 3)
    #     3: {
    #         'n2': {'WindEnergyIn1e51erg': '0.9'},
    #         'n1': {'WindEnergyIn1e51erg': '1.8'},
    #         '0': {'WindEnergyIn1e51erg': '3.6'},
    #         '1': {'WindEnergyIn1e51erg': '7.2'},
    #         '2': {'WindEnergyIn1e51erg': '14.4'}
    #     },
    #     # Add more hardcoded values if needed
    # }
    
    # Try to read from file first
    param_file = f"/home/jovyan/camels/param_info_files/CosmoAstroSeed_{simulation}_L25n256_{dataset}.txt"
    param_values = {}
    
    try:
        with open(param_file, 'r') as f:
            # Read all lines in the file
            lines = f.readlines()
            
            # Parse header to get parameter names and column indices
            header = lines[0].strip()
            if header.startswith('#'):
                header = header[1:].strip()
            column_names = header.split()
            
            # Create mapping from parameter name to column index
            param_name_to_col = {name: idx for idx, name in enumerate(column_names)}
            
            # Initialize parameter values dictionary
            for param_idx in param_idx_to_name.keys():
                param_values[param_idx] = {'n2': {}, 'n1': {}, '0': {}, '1': {}, '2': {}}
            
            # For each variation line, extract the relevant parameter value
            for line in lines[1:]:
                parts = line.strip().split()
                if not parts or len(parts) < 2:
                    continue
                
                name = parts[0]
                if not name.startswith(f"{dataset}_p"):
                    continue
                
                # Extract parameter index and variation from name
                # e.g., "1P_p5_n2" -> param_idx=5, variation="n2"
                try:
                    param_part = name.split('_p')[1]
                    param_idx = int(param_part.split('_')[0])
                    variation = param_part.split('_')[1]
                    
                    if param_idx not in param_idx_to_name or variation not in ['n2', 'n1', '0', '1', '2']:
                        continue
                    
                    # Get the parameter name and its column
                    param_name = param_idx_to_name[param_idx]
                    if param_name not in param_name_to_col:
                        continue
                    
                    col_idx = param_name_to_col[param_name]
                    if col_idx < len(parts):
                        # Initialize the dictionary if needed
                        if param_idx not in param_values:
                            param_values[param_idx] = {'n2': {}, 'n1': {}, '0': {}, '1': {}, '2': {}}
                        if variation not in param_values[param_idx]:
                            param_values[param_idx][variation] = {}
                            
                        param_values[param_idx][variation][param_name] = parts[col_idx]
                    
                except (ValueError, IndexError):
                    continue
            
        return param_values
    
    except Exception as e:
        print(f"Error reading parameter file {param_file}: {str(e)}")
        # print("Using hardcoded values as fallback.")
        # # Fall back to hardcoded values if file reading fails
        # return hardcoded_values
```

```python
def plot_single_parameter_uvlf(simulation, category, redshift_info, param_name, param_idx, dataset="1P"):
    """
    Create a plot showing UVLF for a single parameter across all its variations.
    
    Args:
        simulation: Name of the simulation
        category: 'attenuated' or 'intrinsic'
        redshift_info: Dictionary with redshift information
        param_name: Parameter name (e.g. 'Omega0')
        param_idx: Parameter index (1-28)
        dataset: Dataset name (default: "1P")
    """
    config = get_config(dataset=dataset, simulation=simulation)
    
    # Get parameter display information
    param_display = param_config.get(param_name, {
        "label": param_name,
        "color": "gray",
        "group": "unknown"
    })
    
    # Read parameter values from file
    param_values = read_parameter_values(dataset, simulation)
    
    # Create color gradient for variations
    base_color = param_display["color"]
    color_obj = mpl.colors.to_rgb(base_color)
    
    # Generate colors with varying lightness
    colors = []
    for i, var in enumerate(variations):
        # Darker to lighter (keeping hue)
        if var == 'n2':
            # Darkest
            factor = 0.5
        elif var == 'n1':
            factor = 0.75
        elif var == '0':
            # Original color
            factor = 1.0
        elif var == '1':
            factor = 1.25
        else:  # var == '2'
            # Lightest
            factor = 1.5
            
        # Ensure we don't exceed RGB bounds
        r = min(1.0, color_obj[0] * factor)
        g = min(1.0, color_obj[1] * factor)
        b = min(1.0, color_obj[2] * factor)
        colors.append((r, g, b))
    
    # Create the figure
    fig, ax = plt.subplots(figsize=(8, 6))
    
    # Get the available filters for this category
    available_filters = config["filters"][category]
    # Use the first filter by default (usually GALEX FUV)
    filter_to_use = available_filters[0]
    filter_system = "GALEX"  # This is based on your directory structure
    
    # Flag to track if we've already plotted the fiducial data
    fiducial_plotted = False
    
    # Plot each variation
    for var_idx, var in enumerate(variations):
        try:
            # Get the data directory for the luminosity function
            data_dir = os.path.join(config["lf_data_dir"][category][filter_system], 
                                  redshift_info['label'])
            
            # For fiducial value (var='0'), always use p1_0
            if var == '0':
                # Look for fiducial data in p1_0 file
                fiducial_filename = f"UVLF_{dataset}_p1_0_{filter_to_use.replace(' ', '_')}_{redshift_info['label']}_{category}.txt"
                fiducial_filepath = os.path.join(data_dir, fiducial_filename)
                
                if os.path.exists(fiducial_filepath):
                    data = pd.read_csv(fiducial_filepath, sep='\t')
                    
                    # Get the actual parameter value for the legend
                    param_value = "fiducial"
                    if param_idx in param_values and '0' in param_values[param_idx] and param_name in param_values[param_idx]['0']:
                        try:
                            val = param_values[param_idx]['0'][param_name]
                            param_value = f"fiducial ({val})"
                        except Exception as e:
                            # Just use "fiducial" if there's an error
                            param_value = "fiducial"
                    
                    ax.errorbar(data['magnitude'], 
                              data['phi'],
                              yerr=data['phi_sigma'],
                              fmt='o-', color=colors[var_idx],
                              markersize=5, capsize=3, alpha=0.8,
                              label=param_value)
                    fiducial_plotted = True
                else:
                    print(f"Fiducial file not found: {fiducial_filepath}")
            else:
                # For non-fiducial variations, use the parameter-specific file
                filename = f"UVLF_{dataset}_p{param_idx}_{var}_{filter_to_use.replace(' ', '_')}_{redshift_info['label']}_{category}.txt"
                filepath = os.path.join(data_dir, filename)
                
                if os.path.exists(filepath):
                    data = pd.read_csv(filepath, sep='\t')
                    
                    # Get the actual parameter value for the legend
                    label = f"{var}"  # Default label
                    if param_idx in param_values and var in param_values[param_idx] and param_name in param_values[param_idx][var]:
                        try:
                            val = param_values[param_idx][var][param_name]
                            label = f"{var} ({val})"
                        except Exception as e:
                            # Just use the variation if there's an error
                            label = f"{var}"
                    
                    ax.errorbar(data['magnitude'], 
                              data['phi'],
                              yerr=data['phi_sigma'],
                              fmt='o-', color=colors[var_idx],
                              markersize=5, capsize=3, alpha=0.8,
                              label=label)
                else:
                    print(f"File not found: {filepath}")
        except Exception as e:
            print(f"Error processing variation {var} for param {param_idx}: {str(e)}")
    
    # Style the plot
    ax.grid(True, alpha=0.3, linestyle='--')

    # Use limits from config if available
    x_min, x_max = config["uvlf_limits"]
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(-5, -2)
    
    # Labels
    ax.set_xlabel('M$_{\\rm{AB}}$')#, fontsize=12)
    ax.set_ylabel('log$_{10}$ φ [Mpc$^{-3}$ mag$^{-1}$]')#, fontsize=12)
    
    # Add parameter information to title
    param_label = param_display["label"]
    param_group = param_display["group"].replace('_', ' ').title()
    
    # title = f"{simulation} UVLF: {param_label} ({param_group})\n{redshift_info['label']}, {category}"
    # ax.set_title(title)#, fontsize=12)
    
    # Add legend with a title showing the parameter being varied
    ax.legend()#fontsize=10)#, title=f"Parameter: {param_name}")
    
    # Ensure the plot directory exists
    plot_dir = config["plots_dir"]["UVLFs"][category]
    os.makedirs(plot_dir, exist_ok=True)
    
    # Save the figure with a specific name following your convention
    output_file = os.path.join(plot_dir, 
                             f"UVLF_{redshift_info['label']}_{category}_{dataset}_p{param_idx}.pdf")
    
    plt.tight_layout()
    plt.savefig(output_file)
    plt.close()
    
    return output_file
```

```python
def plot_single_parameter_color(simulation, category, redshift_info, param_name, param_idx, dataset="1P"):
    """
    Create a plot showing color distribution for a single parameter across all its variations.
    
    Args:
        simulation: Name of the simulation
        category: 'attenuated' or 'intrinsic'
        redshift_info: Dictionary with redshift information
        param_name: Parameter name (e.g. 'Omega0')
        param_idx: Parameter index (1-28)
        dataset: Dataset name (default: "1P")
    """
    config = get_config(dataset=dataset, simulation=simulation)
    
    # Get parameter display information
    param_display = param_config.get(param_name, {
        "label": param_name,
        "color": "gray",
        "group": "unknown"
    })
    
    # Read parameter values from file
    param_values = read_parameter_values(dataset, simulation)
    
    # Create color gradient for variations
    base_color = param_display["color"]
    color_obj = mpl.colors.to_rgb(base_color)
    
    # Generate colors with varying lightness
    colors = []
    for i, var in enumerate(variations):
        # Darker to lighter (keeping hue)
        if var == 'n2':
            # Darkest
            factor = 0.5
        elif var == 'n1':
            factor = 0.75
        elif var == '0':
            # Original color
            factor = 1.0
        elif var == '1':
            factor = 1.25
        else:  # var == '2'
            # Lightest
            factor = 1.5
            
        # Ensure we don't exceed RGB bounds
        r = min(1.0, color_obj[0] * factor)
        g = min(1.0, color_obj[1] * factor)
        b = min(1.0, color_obj[2] * factor)
        colors.append((r, g, b))
    
    # Create the figure
    fig, ax = plt.subplots(figsize=(8, 6))
    
    # Get the first color pair from config
    if len(config["colour_pairs"]) > 0:
        filter1, filter2 = config["colour_pairs"][0]
        # Generate the color pair directory name
        colour_pair = get_colour_dir_name(filter1, filter2)
        # Extract color pair components for title
        colour_components = [filter1.split()[-1], filter2.split()[-1]] if ' ' in filter1 else colour_pair.split('-')
    else:
        # Default to GALEX FUV-NUV if no color pairs defined
        filter1, filter2 = "GALEX FUV", "GALEX NUV"
        colour_pair = "GALEX_FUV-NUV"
        colour_components = ["FUV", "NUV"]
    
    # Flag to track if we've already plotted the fiducial data
    fiducial_plotted = False
    
    # Plot each variation
    for var_idx, var in enumerate(variations):
        try:
            # Use the colour data directory from config
            colour_dir = os.path.join(config["colour_data_dir"][category],
                                    colour_pair,
                                    redshift_info['label'])
            
            # For fiducial value (var='0'), always use p1_0
            if var == '0':
                # Look for fiducial data in p1_0 file
                fiducial_filename = f"Colour_{dataset}_p1_0_{colour_pair}_{redshift_info['label']}_{category}.txt"
                fiducial_filepath = os.path.join(colour_dir, fiducial_filename)
                
                if os.path.exists(fiducial_filepath):
                    # Read the color distribution data
                    data = pd.read_csv(fiducial_filepath, sep='\t')
                    
                    # Get the actual parameter value for the legend
                    param_value = "fiducial"
                    if param_idx in param_values and '0' in param_values[param_idx] and param_name in param_values[param_idx]['0']:
                        try:
                            val = param_values[param_idx]['0'][param_name]
                            param_value = f"fiducial ({val})"
                        except Exception as e:
                            # Just use "fiducial" if there's an error
                            param_value = "fiducial"
                    
                    # Use 'color' or 'colour' column (handle both naming conventions)
                    color_column = 'colour' if 'colour' in data.columns else 'color'
                    if color_column in data.columns and 'distribution' in data.columns:
                        ax.plot(data[color_column],
                               data['distribution'],
                               color=colors[var_idx],
                               linewidth=2, alpha=0.8,
                               label=param_value)
                        fiducial_plotted = True
                else:
                    print(f"Fiducial color file not found: {fiducial_filepath}")
            else:
                # For non-fiducial variations, use the parameter-specific file
                filename = f"Colour_{dataset}_p{param_idx}_{var}_{colour_pair}_{redshift_info['label']}_{category}.txt"
                filepath = os.path.join(colour_dir, filename)
                
                if os.path.exists(filepath):
                    # Read the color distribution data
                    data = pd.read_csv(filepath, sep='\t')
                    
                    # Get the actual parameter value for the legend
                    label = f"{var}"  # Default label
                    if param_idx in param_values and var in param_values[param_idx] and param_name in param_values[param_idx][var]:
                        try:
                            val = param_values[param_idx][var][param_name]
                            label = f"{var} ({val})"
                        except Exception as e:
                            # Just use the variation if there's an error
                            label = f"{var}"
                    
                    # Use 'color' or 'colour' column (handle both naming conventions)
                    color_column = 'colour' if 'colour' in data.columns else 'color'
                    if color_column in data.columns and 'distribution' in data.columns:
                        ax.plot(data[color_column],
                               data['distribution'],
                               color=colors[var_idx],
                               linewidth=2, alpha=0.8,
                               label=label)
                else:
                    print(f"Color file not found: {filepath}")
        except Exception as e:
            print(f"Error processing variation {var} for param {param_idx}: {str(e)}")
    
    # Style the plot
    ax.grid(True, alpha=0.3, linestyle='--')
    
    # Use limits from config if available
    x_min, x_max = config["colour_limits"]
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(0, 2.5)  # Adjust y-limits for color distributions
    
    # Labels
    ax.set_xlabel(f'{colour_components[0]}-{colour_components[1]}')#)#, fontsize=12)
    ax.set_ylabel('Probability Density')#)#, fontsize=12)
    
    # Add parameter information to title
    param_label = param_display["label"]
    param_group = param_display["group"].replace('_', ' ').title()
    
    # title = f"{simulation} Colors: {param_label} ({param_group})\n{redshift_info['label']}, {category}"
    # ax.set_title(title)#)#, fontsize=12)
    
    # Add legend with a title showing the parameter being varied
    ax.legend()#fontsize=10)#, title=f"Parameter: {param_name}")
    
    # Ensure the plot directory exists
    plot_dir = config["plots_dir"]["colours"][category]
    os.makedirs(plot_dir, exist_ok=True)
    
    # Save the figure with a specific name following your convention
    output_file = os.path.join(plot_dir, 
                             f"Color_{colour_pair}_{redshift_info['label']}_{category}_{dataset}_p{param_idx}.pdf")
    
    plt.tight_layout()
    plt.savefig(output_file)
    plt.close()
    
    return output_file
```

```python
def main(dataset="1P"):
    """Main function to generate all plots."""
    # Process each simulation
    for simulation in simulations:
        config = get_config(dataset=dataset, simulation=simulation)
        print(f"Processing {simulation}...")
        
        # Create mapping from parameter name to index
        param_name_to_idx = {}
        for param_name, param_info in param_config.items():
            if "index" in param_info:
                param_name_to_idx[param_name] = param_info["index"]
        
        # Process each category (attenuated/intrinsic)
        for category in ['attenuated', 'intrinsic']:
            print(f"  Category: {category}")
            
            # Process each redshift
            for snap, redshift_info in config["redshift_values"].items():
                print(f"    Redshift: {redshift_info['label']}")
                
                # Create directories for plots if they don't exist
                uvlf_dir = config["plots_dir"]["UVLFs"][category]
                os.makedirs(uvlf_dir, exist_ok=True)
                
                color_dir = config["plots_dir"]["colours"][category]
                os.makedirs(color_dir, exist_ok=True)
                
                # First check if fiducial file exists
                filter_system = "GALEX"  # Based on your directory structure
                filter_to_use = config["filters"][category][0]  # Use first filter
                
                # Check fiducial UVLF file
                data_dir = os.path.join(config["lf_data_dir"][category][filter_system],
                                      redshift_info['label'])
                fiducial_file = f"UVLF_{dataset}_p1_0_{filter_to_use.replace(' ', '_')}_{redshift_info['label']}_{category}.txt"
                fiducial_path = os.path.join(data_dir, fiducial_file)
                fiducial_uvlf_exists = os.path.exists(fiducial_path)
                
                # Check fiducial color file
                if len(config["colour_pairs"]) > 0:
                    filter1, filter2 = config["colour_pairs"][0]
                    colour_pair = get_colour_dir_name(filter1, filter2)
                else:
                    colour_pair = "GALEX_FUV-NUV"
                
                colour_dir = os.path.join(config["colour_data_dir"][category],
                                        colour_pair,
                                        redshift_info['label'])
                fiducial_color_file = f"Colour_{dataset}_p1_0_{colour_pair}_{redshift_info['label']}_{category}.txt"
                fiducial_color_path = os.path.join(colour_dir, fiducial_color_file)
                fiducial_color_exists = os.path.exists(fiducial_color_path)
                
                if not fiducial_uvlf_exists:
                    print(f"    WARNING: Fiducial UVLF file not found at {fiducial_path}")
                
                if not fiducial_color_exists:
                    print(f"    WARNING: Fiducial color file not found at {fiducial_color_path}")
                
                # Process each parameter with an index in param_config
                for param_name, param_idx in param_name_to_idx.items():
                    print(f"      Parameter {param_idx}: {param_name}")
                    
                    # For each parameter, check if at least one of the variation files exists
                    uvlf_variation_exists = False
                    color_variation_exists = False
                    
                    # Check for at least one variation file existing for UVLF
                    for var in ['n2', 'n1', '1', '2']:  # Skip '0' as we're using p1_0 for that
                        test_file = f"UVLF_{dataset}_p{param_idx}_{var}_{filter_to_use.replace(' ', '_')}_{redshift_info['label']}_{category}.txt"
                        test_path = os.path.join(data_dir, test_file)
                        if os.path.exists(test_path):
                            uvlf_variation_exists = True
                            break
                    
                    # Check for at least one variation file existing for colors
                    for var in ['n2', 'n1', '1', '2']:  # Skip '0' as we're using p1_0 for that
                        test_file = f"Colour_{dataset}_p{param_idx}_{var}_{colour_pair}_{redshift_info['label']}_{category}.txt"
                        test_path = os.path.join(colour_dir, test_file)
                        if os.path.exists(test_path):
                            color_variation_exists = True
                            break
                    
                    # Plot UVLF if fiducial and at least one variation file exists
                    if fiducial_uvlf_exists and uvlf_variation_exists:
                        try:
                            uvlf_file = plot_single_parameter_uvlf(
                                simulation, category, redshift_info, param_name, param_idx, dataset
                            )
                            print(f"        UVLF plot saved: {uvlf_file}")
                        except Exception as e:
                            print(f"        Error creating UVLF plot for {param_name}: {str(e)}")
                    else:
                        print(f"        Skipping UVLF plot for {param_name}: Missing required files")
                    
                    # Plot color distribution if fiducial and at least one variation file exists
                    if fiducial_color_exists and color_variation_exists:
                        try:
                            color_file = plot_single_parameter_color(
                                simulation, category, redshift_info, param_name, param_idx, dataset
                            )
                            print(f"        Color plot saved: {color_file}")
                        except Exception as e:
                            print(f"        Error creating color plot for {param_name}: {str(e)}")
                    else:
                        print(f"        Skipping color plot for {param_name}: Missing required files")

```
